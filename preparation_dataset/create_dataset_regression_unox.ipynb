```python
import pandas as pd
import os
from google.colab import files
import io
import random
```

```python
uploaded= files.upload()
```

```python
df= pd.read_csv(io.BytesIO(uploaded['training_df.csv']))
```

```python
df_unox= pd.read_csv(io.BytesIO(uploaded['complete_df.csv']))
```

```python
df_unox_filtered= df_unox[df_unox['operator'].str.contains("Uno", case= False, na= False)]
id_list = df_unox_filtered['id'].tolist()
num_row_unox= len(df_unox_filtered)
print(num_row_unox)
```

```python
id_list = df_unox_filtered['id'].tolist()
print("List of IDs:\n", id_list)
```

```python
test_ids = random.sample(id_list, 56)
df['split'] = 'train'
df.loc[df['id'].isin(test_ids), 'split'] = 'test'
non_test_idx = df[~df['id'].isin(test_ids)].index
val_size = int(0.15 * len(df))  # 15% validation set
val_indices = random.sample(list(non_test_idx), val_size)
df.loc[val_indices, 'split'] = 'validation'
```
